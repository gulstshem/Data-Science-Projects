## B. Random Search - randomly decide which parameters will have which value
### B.1.  Setting the random range and parameters

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 3, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [2, 3, 4, 5]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

### B.2. Seting the greed and view of the random results

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}


print(random_grid)
```

### B.3. Running the Rnadom forest model iterating with the greed selected (randomly) parameters

```python
rf = RandomForestRegressor()

rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 50, cv = 10, 
                               verbose=2, random_state=2242, n_jobs = -1)

# Fit the random search model
rf_random.fit(X_train, y_train)
```

## C. Results
### C.1. Best parameters found by the model iteration:

```python
rf_random.best_params_
```

### C.2. Evaluation function

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mae = 100 * np.mean(errors)
    print('Model Performance')
    print('Mean Absolute Error: {:0.4f}'.format(np.mean(errors)))
    return mae
```

### C.3. Run base model - befor the finetuning 


```python
base_model = RandomForestRegressor(n_estimators = 1000, random_state = 4)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
```

# Base model Mean Absolut Error = 0.4782 !!!
### C.4. Compare fine-tunned model


```python
best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)
```

# Finetuning resutl - Mean Absolut Error = 0.4743 !!!

### C.5. Improvmet persentage after fine tuning:

```python
print('Improvement of {:0.2f}%.'.format( 100 * (base_accuracy - random_accuracy) / base_accuracy))
```
